# "Volcano Eruptions"

```python
import numpy as np
import pandas as pd
```

```python
volc = pd.read_csv('Datasets\Input\eruptions.csv', sep=',')
```

```python
volc
```

```python
volc['eruption_category'].value_counts()
```

```python
volc.info()
```

```python
# We set datetime properties for a new column 'Start Date': 

volc.rename(columns={'start_year' : 'Year',
                   'start_month' : 'Month',
                   'start_day' : 'Day'}, inplace=True)


volc ['Start Date'] = pd.to_datetime (volc [["Year", "Month", "Day"]],  errors = 'coerce')
```

```python
# And for 'End Date': 

volc = volc.drop (['Year', 'Month', 'Day'], axis=1)

volc.rename(columns={'end_year' : 'Year',
                   'end_month' : 'Month',
                   'end_day' : 'Day'}, inplace=True)


volc ['End Date'] = pd.to_datetime (volc [["Year", "Month", "Day"]],  errors = 'coerce')
```

```python
volc = volc.drop (['Year', 'Month', 'Day'], axis=1)
```

```python
volc.columns
```

```python
# Rearrange columns order:

volc = volc [['volcano_number', 'volcano_name', 'eruption_number',
       'eruption_category', 'Start Date',
       'End Date', 'area_of_activity', 'vei',
       'evidence_method_dating', 'latitude', 'longitude']]
```

```python
# Final view of finished dataframe

volc
```

```python
volc.to_csv ('Datasets\Output\Volcano-Database.csv')
```
