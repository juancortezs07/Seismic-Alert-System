- ## Unitated States API

Opening the url as a json file

```python
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2014-01-01&endtime=2014-01-02&minlatitude=24.911795455444313&minlongitude=-124.65301770531302&maxlatitude=49.02924641916901&maxlongitude=-66.95789388605114&mindepth=-1000&maxdepth=2000'
json_obj = urlopen(url)
data = json.load(json_obj)
```

```python
df_api_us.rename(columns={'properties.time' : 'time',
                          'properties.place' : 'place',
                          'properties.mag' : 'mag',
                          'properties.magType' : 'magType',
                          'geometry.coordinates' : 'coordinates'}, inplace=True)
```

```python
df_api_us.isna().sum()*100/len(df_api_us) # Checking for nan values
```

I need the coordinates split into latitude and longitude

```python
df_api_us.time = pd.to_datetime(df_api_us.time, unit='ms') # Converting the time column into datetime
```

Final Dataframe

- ## Japan API

### I import the pandas_read_xml library

```python
import pandas_read_xml as pdx
```

### Convert to dataframe the xml data

```python
df = pdx.read_xml('http://service.iris.edu/fdsnws/event/1/query?includeallmagnitudes=TRUE&magnitudetype=all&minlatitude=30.180889170292048&minlongitude=129.20644141355123&maxlatitude=45.540717058168504&maxlongitude=146.68542509079882&starttime=2021-01-01T00:00:00&endtime=2022-01-01T00:00:00&format=xml')
```

### Transform that data into a json object

```python
json_obj = df.to_json()
json_obj
```

### Give the json object a json format

```python
json_format = json.loads(json_obj)
json_format
```

### I normalize the json data into another dataframe

```python
data = pd.json_normalize(json_format)
```

```python
data.keys() # I check for the key with the values I'm looking for
```

```python
df_events = data['q:quakeml.0.eventParameters.event'] # Make another dataframe with just that field values
```

```python
df_events
```

### I put those values into a list of events

```python
events = []
for entry in df_events:
    for l in entry:
        events.append(l)
```

```python
events
```

### The dictionaries path to the values we are interested in

```python
events[0]
```

```python
events[0]['type'] # type
events[0]['description']['text'] # description
events[0]['origin']['time']['value'] # time
events[0]['origin']['creationInfo']['author'] # author
events[0]['origin']['latitude']['value'] # latitude
events[0]['origin']['longitude']['value'] # longitude
events[0]['origin']['depth']['value'] # depth
events[0]['magnitude']['mag']['value'] # magnitude
events[0]['magnitude']['type'] # magnitude type
```

### Function to get data from nested dictionay

```python
from functools import reduce
def deep_get(dictionary, keys, default=None):
    return reduce(lambda d, key: d.get(key, default) if isinstance(d, dict) else default, keys.split("."), dictionary)
```

```python
print(deep_get(events[0], "origin.time.value"))  
```

### Getting the nested data from every dictionary to one dictionary

```python
events_values = {'type' : [], 
                'place' : [], 
                'time' : [],
                'author' : [],
                'latitude' : [],
                'longitude' : [],
                'depth' : [],
                'mag' : [],
                'magType' : []}
                
for event in events:
    events_values['type'].append(deep_get(event, 'type')) 
    events_values['place'].append(deep_get(event, 'description.text')) 
    events_values['time'].append(deep_get(event, 'origin.time.value')) 
    events_values['author'].append(deep_get(event, 'origin.creationInfo.author')) 
    events_values['latitude'].append(deep_get(event, 'origin.latitude.value')) 
    events_values['longitude'].append(deep_get(event, 'origin.longitude.value')) 
    events_values['depth'].append(deep_get(event, 'origin.depth.value')) 
    events_values['mag'].append(deep_get(event, 'magnitude.mag.value')) 
    events_values['magType'].append(deep_get(event, 'magnitude.type'))
```

### Converting the dictionary into a dataframe

```python
df_api_japan = pd.DataFrame(events_values)
```

```python
df_api_japan # Dataframe ready 
```

# Transformations of the dataframe

I check the dtype of the columns

```python
df_api_japan.info() # There are all strings
```

I change the time column to datetime

```python
df_api_japan['time'] = pd.to_datetime(df_api_japan['time'])#, unit='ns')
```

All of the numeric fields need to be convert into float

```python
df_api_japan.latitude = df_api_japan.latitude.astype(float)
df_api_japan.longitude = df_api_japan.longitude.astype(float)
df_api_japan.depth = df_api_japan.depth.astype(float)
df_api_japan.mag = df_api_japan.mag.astype(float)
```

Now I select the columns I need and re-order them

```python
df_api_japan = df_api_japan[['time', 'place', 'mag', 'magType', 'depth', 'latitude', 'longitude']]
```

I'll normalize the place values

```python
df_api_japan.place = df_api_japan.place.str.title()
```

Final Dataframe

```python
df_api_japan
```
