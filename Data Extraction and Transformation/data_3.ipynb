```python
data_3['Class_Mag'] = pd.cut(
    x=data_3["EQ Primary"],
    bins=[0, 2.9, 3.9,4.9,5.9,6.9,7.9, np.inf],
    labels=["Very Minor", "Minor", "Light","Moderate","Strong","Major","Great"],
)
```

```python
data_3.drop(['Mw Magnitude','Ms Magnitude','Mb Magnitude', 'Ml Magnitude','MFA Magnitude','Unknown Magnitude','State','Region code','Earthquake : Missing','Earthquake : Missing Description','Total Effects : Missing','Total Effects : Missing Description'], axis=1,inplace=True)
```

```python
lat = []
lon = []


for row in data_3['Coordinates']:
   
    try:
        
        lat.append(row.split(',')[0])
       
        lon.append(row.split(',')[1])
    
    except:
       
        lat.append(np.NaN)
        
        lon.append(np.NaN)


data_3['latitude'] = lat
data_3['longitude'] = lon
```

```python
data_3.drop(['Coordinates'], axis=1)
```

```python
data_3['Earthquake : Deaths'] = data_3['Earthquake : Deaths'].fillna(0)
```

```python
data_3['Earthquake : Deaths'].head()
```

```python
data_3['Earthquake : Deaths Description']=data_3['Earthquake : Deaths Description'].astype(str)
```

```python
data_3['tmp_table']=''
conditions=[(data_3['Earthquake : Deaths Description'].str.contains('50')),(data_3['Earthquake : Deaths Description'].str.contains('100')),(data_3['Earthquake : Deaths Description'].str.contains('1000'),)]
choices=[15,75,500]
data_3['tmp_table']=np.select(conditions,choices)
```

```python
data_3['Earthquake : Deaths'] = np.where((data_3['Earthquake : Deaths'] == 0), data_3['tmp_table'],data_3['Earthquake : Deaths'])
```

```python
del data_3['tmp_table']
```

```python
data_3['Earthquake : Damage (in M$)'] = data_3['Earthquake : Damage (in M$)'].fillna(0)
```

```python
data_3['Earthquake : Damage Description']=data_3['Earthquake : Damage Description'].astype(str)
```

```python
data_3['tmp_table']=''
conditions=[(data_3['Earthquake : Damage Description'].str.contains('less')),(data_3['Earthquake : Damage Description'].str.contains('MODERATE')),(data_3['Earthquake : Damage Description'].str.contains('SEVERE')),(data_3['Earthquake : Damage Description'].str.contains('EXTREME'))]
choices=[np.random.uniform(0,1),np.random.uniform(1,5),np.random.uniform(5,25),np.random.uniform(25,40000)]
data_3['tmp_table']=np.select(conditions,choices)
```

```python
data_3['Earthquake : Damage (in M$)'] = np.where((data_3['Earthquake : Damage (in M$)'] == 0), data_3['tmp_table'],data_3['Earthquake : Damage (in M$)'])
```

```python
data_3.to_excel(r"data_3_final.xlsx", encoding='utf-8', index=False)
```
