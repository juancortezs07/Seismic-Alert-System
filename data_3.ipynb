```python
import pandas as pd
import numpy as np
```

```python
datachile = pd.read_csv('monetary_losses_data.csv', sep=';')
```

```python
datachile.head()
```

```python
datachile['Date']=pd.to_datetime(datachile[["Year", "Month", "Day"]],  errors = 'coerce')
```

```python
datachile.info()
```

```python
datachile.nunique()
```

```python
data_3=datachile.loc[(datachile['Country']=='CHILE')| (datachile['Country']=='USA') | (datachile['Country']=='JAPAN')]
```

```python
data_3.info()
```

```python
data_3[' ID Earthquake'].head()
```

```python
data_3.loc[data_3[' ID Earthquake']==10036,['Focal Depth']]=2
data_3.loc[data_3[' ID Earthquake']==7614,['Focal Depth']]=5
data_3.loc[data_3[' ID Earthquake']==5538,['Focal Depth']]=20
data_3.loc[data_3[' ID Earthquake']==9813,['Focal Depth']]=25
data_3.loc[data_3[' ID Earthquake']==4248,['Focal Depth']]=20
```

```python
data_3.loc[data_3['EQ Primary'].notna()]
```

```python
data_3.loc[data_3['Focal Depth'].notna()]
```

```python
data_3['Class_Depth'] = pd.cut(
    x=data_3["Focal Depth"],
    bins=[0, 49, 300, np.inf],
    labels=["Shallow", "Intermediate", "Deep"],
)
```

```python
data_3['Class_Mag'] = pd.cut(
    x=data_3["EQ Primary"],
    bins=[0, 2.9, 3.9,4.9,5.9,6.9,7.9, np.inf],
    labels=["Very Minor", "Minor", "Light","Moderate","Strong","Major","Great"],
)
```

```python
data_3.drop(['Mw Magnitude','Ms Magnitude', 'Ml Magnitude','MFA Magnitude','Unknown Magnitude','State','Region code','Earthquake : Missing','Earthquake : Missing Description','Total Effects : Missing','Total Effects : Missing Description'], axis=1)
```

```python
lat = []
lon = []


for row in data_3['Coordinates']:
   
    try:
        
        lat.append(row.split(',')[0])
       
        lon.append(row.split(',')[1])
    
    except:
       
        lat.append(np.NaN)
        
        lon.append(np.NaN)


data_3['latitude'] = lat
data_3['longitude'] = lon
```

```python
data_3.drop(['Coordinates'], axis=1)
```

```python
data_3.to_excel(r"data_3_final.xlsx", encoding='utf-8', index=False)
```
